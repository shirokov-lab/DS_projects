# Mercedes-Benz Greener Manufacturing

## Description

Goal is to predict the time it will take to pass the testing phase. The dataset represents various permutations of the characteristics of Mercedes-Benz vehicles. Reducing the algorithm run time can also help reduce carbon dioxide emissions without compromising Daimler's standards.

The dataset contains an anonymized set of variables (user-defined functions) in a Mercedes vehicle. For example, a variable could be 4WD, it could be an added air suspension, or a head-up display.

y is the variable to be predicted, this is the time (in seconds) it took for the car to be tested for each variable

Variables containing letters are categorical. Variables with 0/1 are of binary type.

### 2.5 Numerical data

```python
num = train.dtypes[train.dtypes=='int64'].index[1:]
num
```

We have a set of numeric variables, where the value is set to 1 or 0, so there is no need to do huge analysis. In this case, we should be interested in whether the value of the indicators changes within the variables, for this we examine the variance of these variables with var() function, and select only those where the variance is zero.

```python
nan_num = []
for i in num:
    if (train[i].var()==0):
        print(i, train[i].var())
        nan_num.append(i)
```

We received several such variables, we can remove them from the analysis, since they will not affect the target y anywayso we can increase the performance of the algorithm.

```python
train = train.drop(columns=nan_num, axis=1)
```

### 2.6 Correlation analysis

To make a correlation analysis for categorical variables we need to transform these variables using LabelEncoder(), when converting the values into a binary form, we will not be able to track the relationship of a particular variable + we must take into account the test set, since its values will participate when finding a target

```python
for i in obj:
    le = LabelEncoder()
    le.fit(list(train[i].values) + list(train[i].values))
    train[i] = le.transform(list(train[i].values))
```

```python
train[obj].head()
```

```python
corr = train[train.columns[1:10]].corr()

fig,ax = plt.subplots(figsize=(7,6))
sns.heatmap(corr, vmax=.7, square=True,annot=True);
```

Among the categorical variables, we did not find a direct corr with the target y

```python
threshold = 1

corr_all = train.drop(columns=obj, axis=1).corr()
corr_all.loc[:,:] =  np.tril(corr_all, k=-1) 
```

```python
already_in = set()
result = []
for col in corr_all:
    perfect_corr = corr_all[col][corr_all[col] == threshold ].index.tolist()
    if perfect_corr and col not in already_in:
        already_in.update(set(perfect_corr))
        perfect_corr.append(col)
        result.append(perfect_corr)
```

```python
result
```

When analyzing numerical variables it was found that some of them have a direct correlation with others, so in order to avoid multicollinearity we can remove the variables with correlation 1 (leave single one of the group), or use regularization so that the algorithm does it in automatic mode.
