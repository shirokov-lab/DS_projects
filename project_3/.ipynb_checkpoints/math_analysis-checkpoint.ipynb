```python
def boxploter(df):
    '''Takes a dataframe and builds subplots for each parameter in three rows for compactness'''
    f, axes = plt.subplots(round(len(info_dic)/3), 3, figsize=(25, 100))
    n = 0
    sns.set(font_scale=1.5)
    for col, info in info_dic.items():
        i, j = divmod(n, 3)
        sns.boxplot(x=df[col], y=df['score'], ax=axes[i, j]
                    ).text(-0.5, 110, info, size='medium')
        n = n+1
    # plt.tight_layout()
    plt.show()
```

```python
boxploter(students)
```

From the visualization, almost every + - parameter affects the score. Some are stronger, some have practically no effect. Now I will find out if there is a statistical difference in the distribution of results by nominative characteristics, using the Student's test. Null hypothesis == distributions of exam results by different parameters ** indistinguishable **:

```python
def get_stat_dif(column):
    cols = students.loc[:, column].value_counts().index
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(students.loc[students.loc[:, column] == comb[0], 'score'],
                     students.loc[students.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.05/len(combinations_all):  # Bonferroni correction
            print('Found statistically significant differences for the parameter -', column)
            model_cols.append(column)
            break
```

```python
for col in info_dic:
    get_stat_dif(col)
```

##### So, in our case, the important parameters that may have an impact on the student's results on the math exam are:

1) whether the student lives in the city or outside the city (t-test)

2) the level of education of the student's mother (t-test)

3) profession of the student's mother (t-test)

4) his failures outside school (t-test)

5) whether the student wants to get higher education (t-test)

6) is the student in a romantic relationship (t-test)

7) how many absences (corr test)

8) age (corr test)

9) is there internet at home (visualization)

```python
model_cols.append('internet')
# I decided to add the Internet as well. believe it can help in identifying a risk group
```

##### Brief interpretation of correlation

**age**:
The younger the student, the more chances for a good exam result. At the same time, there are much more observations of younger students than older ones. It can be roughly said that it is worth paying attention to older students.
**absences**:
Indeed, it can be assumed that the student who skip classes more will write the exam better. But I dont know how the data was collected (maybe a strict teacher gave a questionnaire, few would admit that hes skipping classes), so its still difficult to argue that students who skip school should still be at risk group.
**address**:
Students living in the city do have better results than those who live outside the city. I suppose that it is worth attributing scholars who live outside the city to a potential risk group.
**Medu**:
Students with more educational moms have a better score.
**Mjob**:
The correlation between the student's math score and the work of his mother is a dubious thing, I think. But it should be noted that in the parameters for the further model there are no parameters about the father (Fedu, Fjob), but there are both about the mother. I believe that parameters of this nature should be taken individually, or not taken into account at all (for example, combine all this into 1 variable 'parent').
**failuers**:
Students without extracurricular failures do better in math.
**higher**:
Ambitious students who want to get a higher education write an exam better than those who do not plan to study at a university.
**romantic**:
Students who are not in a relationship score better.
**internet**:
Students with internet access at home take the exam with a higher score than students without internet.

All these parameters together will help to identify the risk group and raise the scores of the math exam.
